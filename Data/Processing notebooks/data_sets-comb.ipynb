```python
import pandas as pd
```

```python
mutations = pd.read_csv('/data/yingfei/cancer_data/mutations_raw_324_final.csv')
print(f'Shape: {mutations.shape}')
mutations.head()
```

```python
cnv_data = pd.read_csv('/data/yingfei/cancer_data/cnv_raw_324_final.csv')
print(f'Shape: {cnv_data.shape}')
cnv_data.head()
```

```python
cell_drug_auc = pd.read_csv('/data/yingfei/cancer_data/cell_drug_auc_final_1111.csv')
print(f'Shape: {cell_drug_auc.shape}')
cell_drug_auc.head()
```

```python
padel_descriptors = pd.read_csv('padel_descriptors.csv')
print(f'Shape: {padel_descriptors.shape}')
padel_descriptors.head()
```

```python
train_cell_lines = pd.read_csv("train_celllines_v1_1111.txt", sep=" ", header = None)
test_cell_lines = pd.read_csv("test_celllines_v1_1111.txt", sep=" ", header = None)
```

```python
### Train Data Set
from  itertools import product
train_celllines = train_cell_lines[0].values
drug_names = cell_drug_auc.columns[1:]
train_data= pd.DataFrame(product(train_celllines, drug_names), columns = ['ARXSPAN_ID', 'DRUG_NAME'])
train_data.head()
```

```python
train_data = train_data.merge(cnv_data, left_on='ARXSPAN_ID', right_on='DepMap-ID', suffixes=(False, False))
train_data.head()
```

```python
train_data = train_data.merge(mutations, left_on='ARXSPAN_ID', right_on='DepMap_ID', suffixes=("_cnv", "_mut"))
train_data.head()
```

```python
train_data = train_data.merge(padel_descriptors, left_on='DRUG_NAME', right_on='Name', how='left', suffixes=(False, False))
train_data.head()
```

```python
train_data = train_data.drop(columns = ['DepMap-ID', 'DepMap_ID','Name'])
train_data['auc'] = 0
for i in range(len(train_data)):
    cell_line = train_data['ARXSPAN_ID'].values[i]
    drug_name = train_data['DRUG_NAME'].values[i]
    train_data.loc[i, 'auc'] = cell_drug_auc.loc[cell_drug_auc.ARXSPAN_ID == cell_line, drug_name].values[0]
print(f'Train Data Shape: {train_data.shape}')
train_data.head()
```

```python
train_data = train_data.loc[train_data['auc'] >= 0]
```

```python
print(train_data.shape)
```

```python
### Test Data Set
test_celllines = test_cell_lines[0].values
drug_names = cell_drug_auc.columns[1:]
test_data= pd.DataFrame(product(test_celllines, drug_names), columns = ['ARXSPAN_ID', 'DRUG_NAME'])
test_data.head()
```

```python
test_data = test_data.merge(cnv_data, left_on='ARXSPAN_ID', right_on='DepMap-ID', suffixes=(False, False))
test_data.head()
```

```python
test_data = test_data.merge(mutations, left_on='ARXSPAN_ID', right_on='DepMap_ID', suffixes=("_cnv", "_mut"))
test_data.head()
```

```python
test_data = test_data.merge(padel_descriptors, left_on='DRUG_NAME', right_on='Name', how='left', suffixes=(False, False))
test_data.head()
```

```python
test_data = test_data.drop(columns = ['DepMap-ID','DepMap_ID','Name'])
test_data['auc'] = 0
for i in range(len(test_data)):
    cell_line = test_data['ARXSPAN_ID'].values[i]
    drug_name = test_data['DRUG_NAME'].values[i]
    test_data.loc[i, 'auc'] = cell_drug_auc.loc[cell_drug_auc.ARXSPAN_ID == cell_line, drug_name].values[0]
print(f'Test Shape: {test_data.shape}')
test_data.head()
```

```python
test_data = test_data.loc[test_data['auc'] >= 0]
```

```python
print(test_data.shape)
```

```python
train_data.to_csv('/data/yingfei/cancer_data/train_data_comb.csv', index = False)
test_data.to_csv('/data/yingfei/cancer_data/test_data_comb.csv', index = False)
```
