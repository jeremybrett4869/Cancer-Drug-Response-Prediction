```python
import pandas as pd 

mutation_df = pd.read_csv('../../data sets/mutations_raw_324_final.csv')
mutation_df.head()
```

```python
cell_drug_df = pd.read_csv('../../data sets/cell_drug_auc_final_1111.csv')
cell_drug_df.head()
```

```python
# LAPATINIB
lapatinib_data = mutation_df.copy()
lapatinib_data['AUC'] = cell_drug_df['LAPATINIB']
lapatinib_data.head()
```

```python
lapatinib_auc = lapatinib_data.AUC.values
len(lapatinib_auc)
```

```python
lapatinib_data[['AUC']].describe()
```

```python
new_lapatinib_data = lapatinib_data.loc[lapatinib_data.AUC >= 0].copy()
new_lapatinib_data
```

```python
import matplotlib.pyplot as plt

plt.hist(new_lapatinib_data['AUC'], 20, color = 'green', histtype = 'bar', rwidth = 0.8)
```

```python
import numpy as np

# binarize the AUC value
threshold_1 = np.quantile(new_lapatinib_data['AUC'], 0.01)
threshold_1
```

```python
new_lapatinib_data['AUC_binary'] = new_lapatinib_data['AUC'].apply(lambda x: 1 if x < threshold_1 else 0)
new_lapatinib_data['AUC_binary'].value_counts()
```

```python
new_lapatinib_data.to_csv('new_lapatinib_data.csv', index = False)
```
