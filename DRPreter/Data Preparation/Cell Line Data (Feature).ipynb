```python
symbol_mapping_dict = {}
not_found_gene_lst = []
for gene, lst in gene_hugo_symbol_dict.items():
    found = False
    for symbol in list(lst.split(", ")):
        if symbol in list(CCLE_mutation.Hugo_Symbol.unique()):
            if gene not in symbol_mapping_dict:
                symbol_mapping_dict[gene] = []
            symbol_mapping_dict[gene].append(symbol)
            found = True
    if not found:
        not_found_gene_lst.append(gene)
        print(gene, "Not Found")
print(len(not_found_gene_lst))   
```

```python
### multiple match
# MICOS10-NBL1 ['MINOS1-NBL1', 'NBL1'] -> MINOS1-NBL1
# CASTOR2 ['GATSL1', 'GATSL2'] -> GATSL2
# NECTIN3 ['PRR3', 'PVRL3'] -> PVRL3
# SEPTIN4 ['C17orf47', 'SEPT4'] -> SEPT4
# NECTIN4 ['PRR4', 'PVRL4'] -> PVRL4
# SEM1 ['C7orf76', 'ECD', 'SHFM1'] -> SHFM1

for gene, lst in symbol_mapping_dict.items():
    if len(lst) > 1:
        print(gene, lst)
```

```python
symbol_mapping_dict_final = {}
for gene, lst in symbol_mapping_dict.items():
    if len(lst) > 1:
        if gene == "MICOS10-NBL1":
            symbol_mapping_dict_final[gene] = "MINOS1-NBL1"
        if gene == "CASTOR2":
            symbol_mapping_dict_final[gene] = "GATSL2"
        if gene == "NECTIN3":
            symbol_mapping_dict_final[gene] = "PVRL3"
        if gene == "SEPTIN4":
            symbol_mapping_dict_final[gene] = "SEPT4"
        if gene == "NECTIN4":
            symbol_mapping_dict_final[gene] = "PVRL4"
        if gene == "SEM1":
            symbol_mapping_dict_final[gene] = "SHFM1"
    else:
        symbol_mapping_dict_final[gene] = lst[0]
```

```python
columns = list(genes_2510)

# define row index names
index_names = drug_response_df.ARXSPAN_ID.values

# create empty dataframe with column names, and row index
data = np.zeros((len(index_names), len(columns))) # initialize all value to be 0
mutation_df = pd.DataFrame(data, columns=columns, index=index_names)
mutation_df.shape
```

```python
count = 0
for i in mutation_df.index.values:
    count += 1
    temp_cell_line_data = CCLE_mutation.loc[CCLE_mutation.DepMap_ID == i].copy()
    for gene in mutation_df.columns.values:
        if gene in temp_cell_line_data.Hugo_Symbol.values:
            mutation_df.loc[i, gene] = 1
        elif gene in symbol_mapping_dict_final:
            if symbol_mapping_dict_final[gene] in temp_cell_line_data.Hugo_Symbol.values:
                mutation_df.loc[i, gene] = 1
    if count % 10 == 0:
        print(count)
```

```python
possible_match_dict = {}
for i in range(len(exp["BUB1B-PAK6"].unique())):
    expression_value = exp["BUB1B-PAK6"].unique()[i]
    if expression_value == 0:
        continue
    print(expression_value)
    cell_line_index = list(exp["BUB1B-PAK6"].values).index(expression_value)
    cell_line = exp.index.values[cell_line_index]
    print(cell_line)
    # print(exp.loc[cell_line,"CASP12"])
    for gene in CCLE_expression_21Q4.columns[1:].values:
        exp_data = CCLE_expression_21Q4.loc[CCLE_expression_21Q4.ID == cell_line, gene].values[0]
        drp_data = exp.loc[cell_line,"BUB1B-PAK6"]
        if np.abs(exp_data - drp_data) < 0.01:
            if gene not in possible_match_dict:
                possible_match_dict[gene] = 0
            possible_match_dict[gene] += 1
    print()
print(len(possible_match_dict))
```
