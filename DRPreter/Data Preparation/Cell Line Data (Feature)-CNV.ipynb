```python
symbol_mapping_dict = {}
not_found_gene_lst = []
for gene in missing_gene_lst:
    if gene not in gene_hugo_symbol_dict:
        print(gene)
        continue
    lst = gene_hugo_symbol_dict[gene]
    found = False
    for symbol in list(lst.split(", ")):
        if symbol in CCLE_genes:
            if gene not in symbol_mapping_dict:
                symbol_mapping_dict[gene] = []
            symbol_mapping_dict[gene].append(symbol)
            found = True
    if not found:
        not_found_gene_lst.append(gene)
        print(gene, "Not Found")
print(len(not_found_gene_lst))   
```

```python
### multiple match
# BUB1B-PAK6 ['PAK5', 'PAK6'] -> PAK6
# H3F3A ['H3-3A', 'H3-3B'] -> H3-3A

for gene, lst in symbol_mapping_dict.items():
    if len(lst) > 1:
        print(gene, lst)
```

```python
symbol_mapping_dict_final = {}
for gene, lst in symbol_mapping_dict.items():
    if len(lst) > 1:
        if gene == "BUB1B-PAK6":
            symbol_mapping_dict_final[gene] = "PAK6"
        if gene == "H3F3A":
            symbol_mapping_dict_final[gene] = "H3-3A"
    else:
        symbol_mapping_dict_final[gene] = lst[0]
```

```python
columns = list(genes_2510)

# define row index names
index_names = drug_response_df.ARXSPAN_ID.values

# create empty dataframe with column names, and row index
data = np.zeros((len(index_names), len(columns))) # initialize all value to be 0
cnv_df = pd.DataFrame(data, columns=columns, index=index_names)
cnv_df.shape
```

```python
CCLE_cnv.columns = list(map(lambda x:x.split(" ")[0], CCLE_cnv.columns))
CCLE_cnv.head()
```

```python
count = 0
for i in cnv_df.index.values:
    count += 1
    for gene in cnv_df.columns.values:
        if gene in CCLE_genes:
            cnv_df.loc[i, gene] = CCLE_cnv.loc[i,gene]
        elif gene in symbol_mapping_dict_final:
            if symbol_mapping_dict_final[gene] in CCLE_genes:
                cnv_df.loc[i, gene] = CCLE_cnv.loc[i,symbol_mapping_dict_final[gene]]
    if count % 10 == 0:
        print(count)
```
